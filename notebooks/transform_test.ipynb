```python
import sys
sys.path.append("/Users/ziminqian/Desktop/ersilia")


import pandas as pd
from transform import Transform


# Load CSV
df = pd.read_csv("output_all.csv", skiprows=1) #fix the label(non numerical row)


extra_cols = df.iloc[:, [0, 1]] #keep columns to add back later


df_numeric = df.drop(df.columns[[0, 1]], axis=1) #drop the first and second column
# df_numeric = df.select_dtypes(include=["number"]) # Drop any rows with non-numeric values




# # Initialize transformer
transformer = Transform(False, True)


# # Run the transformation
df_transformed = transformer.run(df_numeric)


# # Assertions on the transformed DataFrame
assert isinstance(df_transformed, pd.DataFrame)
# assert df_transformed.shape == df.shape
# assert df_transformed.min().min() >= -128
# assert df_transformed.max().max() <= 127


df_final = pd.concat([extra_cols.reset_index(drop=True), df_transformed], axis=1)


df_final.to_csv('new_scaled.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt




df = pd.read_csv("five_transf_output.csv", header=None)


x_values = df.iloc[1:, 0].astype(str)
y_values = df.iloc[1:, 2].astype(float) 




plt.figure(figsize=(10, 4))
plt.plot(x_values, y_values, marker='o')  # optional: marker='o' to show points


plt.xlabel("Compound")
plt.ylabel("ABC")
plt.title("ABC Feature Across 5 Compounds")
plt.xticks(rotation=45, ha='right')  # for better label readability
plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt




df = pd.read_csv("all_transf_output.csv", header=None)


# ABC column
col = df.iloc[:, 2].astype(int) 


# Count occurrences of each value between -127 and 128
value_range = range(-127, 129)
counts = col.value_counts().reindex(value_range, fill_value=0)


# Plot
plt.figure(figsize=(12, 5))
plt.bar(counts.index, counts.values, width=1)
plt.xlabel("Value")
plt.ylabel("Count")
plt.title("Distribution of Scaled/Quantized Values of ABC Feature in Mordred Model")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


df = pd.read_csv("new_scaled.csv", header=None)
col = df.iloc[:, 2].astype(float)


plt.figure(figsize=(10, 5))
sns.kdeplot(col, bw_adjust=3, fill=True)  # bw_adjust controls the smoothness
plt.xlim(-7, 7)  # Limit x-axis from -3 to 3
plt.xlabel("Scaled Value")
plt.ylabel("Density")
plt.title("KDE of Scaled ABC Feature")
plt.tight_layout()
plt.show()
```
